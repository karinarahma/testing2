```python
#export df and image to excel
import requests
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from scrapy import Selector
import pandas as pd
from IPython.core.display import HTML
from io import BytesIO
import xlsxwriter


headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}

productlinks = []

for x in range(1,4):
    r = requests.get(f'https://www.tokopedia.com/search?q=kue%20lapis%20legit&source=universe&st=product&page={x}', headers=headers)
    soup = BeautifulSoup(r.content,'lxml')
    productlist = soup.find_all('div', class_ = 'css-1g20a2m')

    for item in productlist:
        for link in item.find_all('a', href=True):
            productlinks.append(link['href'])


lapislegitlist = []
lapis_images = []
for link in productlinks:
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content,'lxml')
    name = Selector(text = requests.get(link, headers=headers).content).xpath('//h1[@class="css-x7lc0h"]/text()').extract()[0]
    price = soup.find('h3', class_='css-c820vl').text.strip().replace('Rp','').replace('.','')
    location = Selector(text = requests.get(link, headers=headers).content).xpath('//span[@data-testid="lblPDPFooterLastOnline"]/text()')[0].extract()
    shop = Selector(text = requests.get(link, headers=headers).content).xpath('//a[@class="css-xmjuvc"]//text()').extract()[0]
    images = Selector(text = requests.get(link, headers=headers).content).xpath('//img[@class="success fade"]/@src').extract()[0]
    try:
        rating = Selector(text = requests.get(link, headers=headers).content).xpath('//span[@data-testid="lblPDPDetailProductRatingNumber"]/text()').extract()[0]
    except:
        rating = '0.0'
        
    lapis_legit = {'Nama Produk': name,
                   'Harga': price,
                   'Rating': rating,
                   'Lokasi': location,
                   'Nama Toko': shop}
#                   'Gambar': images}

    lapislegitlist.append(lapis_legit)
    lapis_images.append(images)
    print('Saving :' , lapis_legit['Nama Produk'])
    
df3 = pd.DataFrame(lapislegitlist)
df3['Harga'] = df3['Harga'].astype(int)

writer = pd.ExcelWriter('LapisLegit2.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df3.to_excel(writer, sheet_name='Sheet1')

workbook  = writer.book
worksheet = writer.sheets['Sheet1']

worksheet.set_column('G2:G2',12)
worksheet.set_default_row(60)

cell_format = workbook.add_format({'bold': True, 'align': 'center', 'valign': 'top', 'border': 1})
worksheet.write(0,6,'Gambar', cell_format)

#insert images
image_row = 1
image_col = 6
for image in lapis_images:
    image_data = BytesIO(urlopen(image).read())
    worksheet.insert_image(image_row,
                           image_col,
                           image,
                           {'image_data' : image_data, 'x_scale': 0.35, 'y_scale' : 0.35, 'x_offset ': 3.5, 'y_offset' : 3.5})

    #positioning = 1 allows move and size with cells
    image_row += 1


writer.save()
```
